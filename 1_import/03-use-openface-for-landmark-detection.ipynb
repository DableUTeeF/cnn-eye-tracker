```python
import subprocess, os
import pandas as pd
```

```python
# Helper func
def make_dir(newpath):
    if not os.path.exists(newpath):
        os.makedirs(newpath)
        
def extract_lankmark_single(img):
    FNULL = open(os.devnull, 'w')    #use this if you want to suppress output to stdout from the subprocess
    command = OPENFACE_DIR + "FaceLandmarkImg.exe -f \"" + img + "\"" + " -2Dfp -pose"
    subprocess.call(command, stdout=FNULL, stderr=FNULL, shell=False)

def extract_lankmark_multiple(series, p_id, task):
    image_string = ''.join([ " -f \"" + i + "\"" for i in series])
    FNULL = open(os.devnull, 'w')
    output_path = RAW_PATH +"facial_landmark\\"+ p_id +"\\"+task+'\\'
    to_extract = " -2Dfp -pose"
    command = OPENFACE_DIR+"FaceLandmarkImg.exe"+image_string +" -out_dir "+output_path+to_extract
    subprocess.call(command, stdout=FNULL, stderr=FNULL, shell=False)
```

```python
OPENFACE_DIR = "D:\\david_thesis\\OpenFace_exe\\"
IMG_PATH = "D:\\FramesDataset\\"
DATASET_PATH = "D:\\david_thesis\\0_data_lan\\datasets\\"
RAW_PATH = DATASET_PATH + "raw\\"
MST_PATH = DATASET_PATH + "mst\\"
```

```python
df = pd.read_csv(MST_PATH + 'all_img.csv')
df.shape
```

```python
pd.set_option('max_columns', 100)
df.head()
```

```python
step = 10
skip = 1
for p_id in df['participant'].unique():
    for task in df['task'].unique():
        temp = df[(df['participant'] == p_id) & (df['task'] == task)]
        if len(temp) == 0: continue
        temp.index = range(len(temp))
        
        print("Starting for:", p_id, task)
        output_path = RAW_PATH +"facial_landmark\\"+ p_id +"\\"+task+'\\'
        make_dir(output_path)
        for i in range(0, temp.shape[0], step):
            if i % 100 == 0: 
                print("Step:", i)
                print("Number of files in current folder:", len(os.listdir(output_path)))
            extract_lankmark_multiple(temp['frameImageFile'][i:i+step:skip], p_id, task)
```
