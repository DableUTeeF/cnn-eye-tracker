```python
# Load in all dependencies and helper functions from the main folder
import sys
sys.path.append('..//')
from utils import *
```

```python
# Helper functions
def clean_path(path):
    return path.replace("../FramesDataset//", IMG_PATH).replace("/", "\\").replace("david_thesis", "david_xu_thesis")

def process_csv(df):
    df['frameImageFile'] = df['frameImageFile'].map(clean_path)
    df = df[['participant', 'frameImageFile', 'frameTimeEpoch', 'frameNum', 'webGazerX', 'webGazerY', \
               'tobiiLeftScreenGazeX', 'tobiiLeftScreenGazeY', 'tobiiRightScreenGazeX', 'tobiiRightScreenGazeY']]
    return df
```

```python
all_csv = os.listdir(RAW_PATH + '\\gaze_predictions')
all_csv[0]
```

```python
for i in range(len(all_csv)):
    print("Start processing:", all_csv[i])
    if i == 0:
        df = process_csv(pd.read_csv(RAW_PATH + '\\gaze_predictions\\' + all_csv[i]))
    else:
        df = df.append(process_csv(pd.read_csv(RAW_PATH + '\\gaze_predictions\\' + all_csv[i])))
    print("df.shape:", df.shape)
```

```python
# Handle wrong parsing
df = df[df['tobiiLeftScreenGazeX'] != 'tobiiLeftScreenGazeX']
```

```python
# Add new column, the sub folder name:
df['task'] = df['frameImageFile'].str.split('\\').map(lambda x: x[3])
```

```python
# Change column dtypes
for col in list(df)[-4:]:
    df[col] = pd.Series(df[col], dtype = np.float32)
    
print(df.dtypes)
```

```python
# Reindex
df.index = range(len(df))
print(list(df))
```
