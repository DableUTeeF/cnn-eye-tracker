```python
import subprocess, os
import pandas as pd
from collections import defaultdict
```

```python
IMG_PATH = "D:\\FramesDataset\\"
DATASET_PATH = "D:\\david_thesis\\0_data_lan\\datasets\\"
RAW_PATH = DATASET_PATH + "raw\\"
MST_PATH = DATASET_PATH + "mst\\"
FACIAL_PATH = RAW_PATH + "facial_landmark\\"
```

```python
df = defaultdict(list)

# Process first 22
for p in os.listdir(FACIAL_PATH):
    p_folder_path = FACIAL_PATH + p + "\\"
    for task in os.listdir(p_folder_path):
        print("Starting for", p, task)
        task_path = p_folder_path + task + "\\"
        for img_csv in [f for f in os.listdir(task_path) if f.count('csv') > 0]:
            df['id'].append(p + "_||_" + task + "_||_" + img_csv[:-4])
            temp = pd.read_csv(task_path + img_csv)
            for col in temp.columns:
                df[col].append(temp.loc[0, col])
```

```python
df = pd.DataFrame(df)
df.head()
```

```python
for col in df.columns:
    print(col, df[col].nunique())
```

```python
# Clean up column names
df.columns = [c.strip() for c in df.columns]
```

```python
df.to_csv(MST_PATH + "all_facial_landmarks.csv", index = None)
```
