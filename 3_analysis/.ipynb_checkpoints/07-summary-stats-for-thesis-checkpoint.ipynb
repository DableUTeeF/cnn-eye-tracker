```python
# One-time helper func
def pie(col_name, save_fig = True):
    counts = p[col_name].value_counts()
    colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', 'gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
    plt.pie(counts, labels=counts.index.values, colors=colors[:len(counts.index)], autopct='%1.1f%%', startangle=90)
    if save_fig: plt.savefig(EXPORT_PATH + '07-summary-stats-for-thesis//' + col_name + '_pie', dpi = 300)
    plt.show()
```

```python
df = pd.read_csv(MST_PATH + 'all_img_facial_lankmarks.csv')
p = pd.read_csv(MST_PATH + 'participant_characteristics.csv')
label = pd.read_csv(MST_PATH + 'mst_eye_label.csv')
```

```python
print(p.shape)
print(df.shape)
```

# Combining the datasets

```python
print(df.shape)
df = pd.merge(df, label, on = 'id', how = 'inner')
print(df.shape)
```

```python
p.columns = ['participant' if x == 'participant_id' else x for x in p.columns]
p = p[p['participant'].isin(df['participant'].unique())]
```

```python
df = df[['participant', 'frameImageFile', 'webGazerX', 'webGazerY', 'coord_x', 'coord_y', 'task', 'img_name']]
df = pd.merge(df , p, on = 'participant', how = 'left')
df.shape
```

```python
# Settings
p_setting = p['setting'].value_counts()
df_setting = df['setting'].value_counts()
res = pd.DataFrame()
res['Num. Subjects'] = p_setting
res['Percent'] = p_setting/sum(p_setting)
res['Num. Images'] = df_setting
res['Percent1'] = df_setting/sum(df_setting)
res.transpose()
```

```python
# Settings
p_setting = p['gender'].value_counts()
df_setting = df['gender'].value_counts()
res1 = pd.DataFrame()
res1['Num. Subjects'] = p_setting
res1['Percent'] = p_setting/sum(p_setting)
res1['Num. Images'] = df_setting
res1['Percent1'] = df_setting/sum(df_setting)
res1.transpose().join(res.transpose())
```
